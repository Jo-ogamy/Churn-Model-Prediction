```python
df_churn_st = pd.read_excel(r"C:\Users\USER\Desktop\Churn_Content\Churn_Status.xlsx")
df_churn_st.head()
```

```python
print(df_churn_st.info())
```

```python
# Calculate the churn percentage
churn_counts = df_churn_st["ChurnStatus"].value_counts(normalize=True) * 100

# Create a bar plot for churn percentage
plt.figure(figsize=(6, 4))
sns.barplot(x=churn_counts.index, y=churn_counts.values, palette="Set2")

# Add titles and labels
plt.title("ChurnStatus Percentage", fontsize=16)
plt.xlabel("ChurnStatus", fontsize=12)
plt.ylabel("Percentage", fontsize=12)

# Annotating the plot with percentages
for i, value in enumerate(churn_counts.values):
    plt.text(i, value + 1, f"{value:.2f}%", ha="center", fontsize=12)

# Show the plot
plt.show()
```

```python
df = pd.concat([df_cus_demo, df_cus_serv, df_online, df_transac, df_churn_st], axis=1)
df.head()
```

```python
df.drop(columns=["CustomerID", "InteractionID", "TransactionID", "TransactionDate"], inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
# Box Plot: Amount Spent by Marital Status
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='MaritalStatus', y='AmountSpent')
plt.title('Amount Spent by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Amount Spent')
plt.show();
```

```python
# Time Series Plot: Interaction Date vs. Amount Spent
plt.figure(figsize=(10, 6))
df['InteractionDate'] = pd.to_datetime(df['InteractionDate'])
sns.lineplot(data=df, x='InteractionDate', y='AmountSpent')
plt.title('Amount Spent Over Time')
plt.xlabel('Interaction Date')
plt.ylabel('Amount Spent')
plt.xticks(rotation=45);
```

```python
# Grouped bar chart for average spending by income level
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='IncomeLevel', y='AmountSpent', estimator='mean')
plt.title('Average Spending by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Average Amount Spent');
```

```python
# Scatter plot for login frequency vs. amount spent
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='LoginFrequency', y='AmountSpent', hue='IncomeLevel')
plt.title('Login Frequency vs. Amount Spent')
plt.xlabel('Login Frequency')
plt.ylabel('Amount Spent')
plt.show();
```

```python
# Bar Chart: Churn Status by Service Usage
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='ServiceUsage', hue='ChurnStatus')
plt.title('Churn Status by Service Usage')
plt.xlabel('Service Usage')
plt.ylabel('Number of Customers')
plt.legend(title='Churn Status', labels=['Not Churned', 'Churned']);
```

```python
# Bar Chart: Service Usage vs. Amount Spent
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='ServiceUsage', y='AmountSpent', estimator=sum, errorbar=None)
plt.title('Total Amount Spent by Service Usage')
plt.xlabel('Service Usage')
plt.ylabel('Total Amount Spent');
```

```python
# Group Bar Chart for Income Level and Product Category
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='IncomeLevel', hue='ProductCategory')
plt.title('Income Level vs. Product Category')
plt.xlabel('Income Level')
plt.ylabel('Number of Purchases')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left');
```

```python
#Bar Chart: Spending pattern and Churn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='AmountSpent', y='ChurnStatus', estimator=sum, errorbar=None)
plt.title('Spending Pattern and Churn')
plt.xlabel("AmountSpent")
plt.ylabel("ChurnStatus");
```

```python
# Visualization of ChurnStatus by income level
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='IncomeLevel', y='ChurnStatus')
plt.title('Churn Rate by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Churn Rate');
```

```python
# Ploting Heatmap for correlation between age, income level, login frequency, and amount spent
# Convert IncomeLevel to numeric for correlation calculation
income_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
df['IncomeLevelClass'] = df['IncomeLevel'].map(income_mapping)
corr_df = df[['Age', 'IncomeLevelClass', 'LoginFrequency', 'AmountSpent']].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show();
```
